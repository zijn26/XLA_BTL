Hiá»ƒn thá»‹ Ä‘áº·c trÆ°ng Ä‘áº« trÃ­ch xuáº¥t Ä‘Æ°á»£c tá»« áº£nh


```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np
import cv2

# 1. XÃ¢y dá»±ng mÃ´ hÃ¬nh CNN Ä‘Æ¡n giáº£n
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(64,64,1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2))
])

# 2. Chuáº©n bá»‹ áº£nh Ä‘áº§u vÃ o (vÃ­ dá»¥ áº£nh hÃ¬nh há»c)
img = cv2.imread("shape.png", cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (64,64))
img = img / 255.0
img = np.expand_dims(img, axis=(0,-1))  # (1,64,64,1)

# 3. Táº¡o model trung gian Ä‘á»ƒ láº¥y output cá»§a tá»«ng layer
layer_outputs = [layer.output for layer in model.layers if 'conv' in layer.name or 'max_pooling2d' in layer.name]
activation_model = models.Model(inputs=model.input, outputs=layer_outputs)

# 4. Cháº¡y áº£nh qua model Ä‘á»ƒ láº¥y feature maps
activations = activation_model.predict(img)

# 5. Hiá»ƒn thá»‹ feature maps cá»§a tá»«ng layer
for layer_name, layer_activation in zip([layer.name for layer in layer_outputs], activations):
    n_features = layer_activation.shape[-1]  # sá»‘ kÃªnh Ä‘áº·c trÆ°ng
    size = layer_activation.shape[1]         # kÃ­ch thÆ°á»›c feature map

    print(f"Layer {layer_name}: {n_features} feature maps, size {size}x{size}")

    # Hiá»ƒn thá»‹ 6 feature maps Ä‘áº§u tiÃªn
    display_grid = np.zeros((size, size * 6))
    for i in range(6):
        feature_map = layer_activation[0, :, :, i]
        feature_map -= feature_map.mean()
        feature_map /= (feature_map.std() + 1e-5)
        feature_map *= 64
        feature_map += 128
        feature_map = np.clip(feature_map, 0, 255).astype('uint8')
        display_grid[:, i*size:(i+1)*size] = feature_map

    plt.figure(figsize=(12, 12))
    plt.title(layer_name)
    plt.imshow(display_grid, aspect='auto', cmap='viridis')
    plt.show()
```

TrÃ­ch chá»n Ä‘áº·c trÆ°ng

```python

import cv2
import numpy as np

def extract_basic_features(image):
    # 1. Chuyá»ƒn sang grayscale náº¿u áº£nh lÃ  RGB
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) if len(image.shape) == 3 else image

    # 2. Nhá»‹ phÃ¢n hÃ³a áº£nh (threshold)
    _, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)

    # 3. TÃ¬m contour
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    features = []

    for cnt in contours:
        # Diá»‡n tÃ­ch vÃ  chu vi
        area = cv2.contourArea(cnt)
        perimeter = cv2.arcLength(cnt, True)

        if perimeter == 0:  # trÃ¡nh chia cho 0
            continue

        # Circularity
        circularity = 4 * np.pi * (area / (perimeter * perimeter))

        # Xáº¥p xá»‰ Ä‘a giÃ¡c Ä‘á»ƒ Ä‘áº¿m sá»‘ Ä‘á»‰nh
        epsilon = 0.02 * perimeter
        approx = cv2.approxPolyDP(cnt, epsilon, True)
        vertices = len(approx)

        # Bounding box Ä‘á»ƒ tÃ­nh aspect ratio
        x, y, w, h = cv2.boundingRect(cnt)
        aspect_ratio = float(w) / h if h != 0 else 0

        # Hu Moments
        hu = cv2.HuMoments(cv2.moments(cnt)).flatten()

        # Gom táº¥t cáº£ Ä‘áº·c trÆ°ng vÃ o dict
        feature_vector = {
            "area": area,
            "perimeter": perimeter,
            "circularity": circularity,
            "vertices": vertices,
            "aspect_ratio": aspect_ratio,
            "hu_moments": hu.tolist()
        }

        features.append(feature_vector)

    return features
```

GhÃ©p dáº·c trÆ°ng trÃ­ch thá»§ cÃ´ng vÃ o bá»• sung cho Ä‘áº·c trÆ°ng phá»©c táº¡p mÃ  CNN trÃ­ch ra Ä‘Æ°á»£c
CNN vá»‘n Ä‘Ã£ tá»± trÃ­ch chá»n Ä‘áº·c trÆ°ng
CNN dÃ¹ng Conv2D + Pooling Ä‘á»ƒ tá»± Ä‘á»™ng há»c Ä‘áº·c trÆ°ng tá»« áº£nh (cáº¡nh, gÃ³c, texture, hÃ¬nh dáº¡ng).

VÃ¬ váº­y, náº¿u báº¡n chá»‰ dÃ¹ng CNN thuáº§n thÃ¬ khÃ´ng cáº§n Ä‘áº·c trÆ°ng thá»§ cÃ´ng (area, perimeter, Hu momentsâ€¦).

ðŸ”¹ Khi nÃ o nÃªn dÃ¹ng Ä‘áº·c trÆ°ng thá»§ cÃ´ng vá»›i CNN?
CÃ³ 2 hÆ°á»›ng káº¿t há»£p:

1. DÃ¹ng riÃªng biá»‡t (khÃ´ng káº¿t há»£p)
Náº¿u dá»¯ liá»‡u Ã­t, mÃ´ hÃ¬nh CNN khÃ³ há»c â†’ báº¡n cÃ³ thá»ƒ dÃ¹ng Ä‘áº·c trÆ°ng thá»§ cÃ´ng + SVM/RandomForest.

ÄÃ¢y lÃ  cÃ¡ch truyá»n thá»‘ng, dá»… giáº£i thÃ­ch, nhÆ°ng khÃ´ng táº­n dá»¥ng sá»©c máº¡nh CNN.

2. Káº¿t há»£p (Hybrid model)
Báº¡n cÃ³ thá»ƒ ghÃ©p Ä‘áº·c trÆ°ng thá»§ cÃ´ng vÃ o táº§ng fully-connected cá»§a CNN.

Quy trÃ¬nh:

áº¢nh â†’ CNN â†’ vector Ä‘áº·c trÆ°ng há»c Ä‘Æ°á»£c (feature embedding).

áº¢nh â†’ trÃ­ch Ä‘áº·c trÆ°ng thá»§ cÃ´ng (area, vertices, Hu momentsâ€¦).

GhÃ©p 2 vector nÃ y láº¡i â†’ Ä‘Æ°a vÃ o Dense layer â†’ Softmax phÃ¢n loáº¡i.

ðŸ‘‰ CÃ¡ch nÃ y gá»i lÃ  feature fusion: CNN há»c Ä‘áº·c trÆ°ng phá»©c táº¡p, cÃ²n Ä‘áº·c trÆ°ng thá»§ cÃ´ng bá»• sung thÃ´ng tin hÃ¬nh há»c rÃµ rÃ ng.

Tiá»n xá»­ lÃ­ áº£nh â‰ÄÃ¢y lÃ  má»™t lá»™ trÃ¬nh tiá»n xá»­ lÃ½ áº£nh (Image Preprocessing) chi tiáº¿t mÃ  báº¡n nÃªn thá»±c hiá»‡n Ä‘á»ƒ chuáº©n bá»‹ dá»¯ liá»‡u Ä‘áº§u vÃ o cho mÃ´ hÃ¬nh CNN nháº±m Nháº­n dáº¡ng hÃ¬nh há»c cÆ¡ báº£n (Äá» tÃ i 6):
Tiá»n xá»­ lÃ½ áº£nh (Image Preprocessing) lÃ  má»™t pháº§n quan trá»ng náº±m giá»¯a Thu nháº­n áº£nh (Image Acquisition) vÃ  Xá»­ lÃ½ chÃ­nh (Computer/CNN).
I. LÃ m sáº¡ch áº£nh vÃ  TÄƒng cÆ°á»ng cháº¥t lÆ°á»£ng (Image Restoration/Enhancement)
Má»¥c tiÃªu lÃ  giáº£m nhiá»…u mÃ  váº«n giá»¯ Ä‘Æ°á»£c biÃªn (Ä‘Æ°á»ng nÃ©t) cá»§a hÃ¬nh khá»‘i.
â€¢ 1. Chuyá»ƒn áº£nh xÃ¡m (Grayscale Conversion): Náº¿u áº£nh Ä‘áº§u vÃ o lÃ  áº£nh mÃ u, báº¡n nÃªn chuyá»ƒn sang áº£nh xÃ¡m (grayscale) Ä‘á»ƒ giáº£m chiá»u dá»¯ liá»‡u, vÃ¬ hÃ¬nh dáº¡ng thÆ°á»ng khÃ´ng phá»¥ thuá»™c vÃ o mÃ u sáº¯c.
â€¢ 2. Loáº¡i bá» nhiá»…u:
    â—¦ Ká»¹ thuáº­t: Sá»­ dá»¥ng Bá»™ lá»c trung vá»‹ (Median Filter).
    â—¦ LÃ½ do: Bá»™ lá»c trung vá»‹ loáº¡i bá» nhiá»…u hiá»‡u quáº£ (Ä‘áº·c biá»‡t lÃ  nhiá»…u muá»‘i tiÃªu - Impulse noise/Salt and pepper noise) vÃ  quan trá»ng hÆ¡n lÃ  báº£o toÃ n Ä‘Æ°á»ng biÃªn (edge) tá»‘t hÆ¡n so vá»›i Bá»™ lá»c trung bÃ¬nh (Mean Filter).
II. TÃ¡ch Ä‘á»‘i tÆ°á»£ng (Segmentation)
Má»¥c tiÃªu lÃ  tÃ¡ch hÃ¬nh khá»‘i ra khá»i ná»n (background) hoáº·c xÃ¡c Ä‘á»‹nh chÃ­nh xÃ¡c Ä‘Æ°á»ng biÃªn cá»§a chÃºng.
â€¢ 3. PhÃ¡t hiá»‡n biÃªn (Edge Detection):
    â—¦ Ká»¹ thuáº­t Æ°u tiÃªn: PhÆ°Æ¡ng phÃ¡p Canny. Canny Ä‘Æ°á»£c coi lÃ  tá»‘i Æ°u vÃ¬ nÃ³ cÃ³ tá»‰ lá»‡ lá»—i tháº¥p vÃ  xÃ¡c Ä‘á»‹nh tá»‘t vá»‹ trÃ­ biÃªn.
â€¢ 4. Thay tháº¿/Bá»• sung: PhÃ¢n ngÆ°á»¡ng (Thresholding): Náº¿u áº£nh cá»§a báº¡n lÃ  hÃ¬nh váº½ Ä‘Æ¡n giáº£n (hai mÃ u sÃ¡ng/tá»‘i rÃµ rá»‡t), phÃ¢n ngÆ°á»¡ng lÃ  má»™t lá»±a chá»n nhanh chÃ³ng.
    â—¦ Ká»¹ thuáº­t: CÃ³ thá»ƒ dÃ¹ng Thuáº­t toÃ¡n Otsu Ä‘á»ƒ tá»± Ä‘á»™ng tÃ¬m ngÆ°á»¡ng tá»‘i Æ°u.
III. Xá»­ lÃ½ hÃ¬nh thÃ¡i há»c (Morphological Processing)
Má»¥c tiÃªu lÃ  tinh chá»‰nh Ä‘Æ°á»ng biÃªn vÃ  hÃ¬nh dáº¡ng Ä‘Ã£ tÃ¡ch (thÆ°á»ng trÃªn áº£nh nhá»‹ phÃ¢n) Ä‘á»ƒ chuáº©n hÃ³a Ä‘áº§u vÃ o cho CNN.
â€¢ 5. Tinh chá»‰nh hÃ¬nh dáº¡ng:
    â—¦ PhÃ©p má»Ÿ (Opening): GiÃºp lÃ m trÆ¡n cÃ¡c Ä‘Æ°á»ng gÃ£y khÃºc, lÃ m má»‹n ranh giá»›i, cáº¯t Ä‘á»©t cÃ¡c dáº£i eo háº¹p khÃ´ng cáº§n thiáº¿t (tÃ¡ch cÃ¡c váº­t thá»ƒ Ä‘Æ°á»£c káº¿t ná»‘i), vÃ  loáº¡i bá» cÃ¡c máº£nh nhiá»…u nhá».
    â—¦ PhÃ©p Ä‘Ã³ng (Closing): GiÃºp káº¿t ná»‘i cÃ¡c Ä‘á»‘i tÆ°á»£ng bá»‹ ngáº¯t káº¿t ná»‘i vÃ  loáº¡i bá» cÃ¡c khoáº£ng trá»‘ng nhá» trÃªn áº£nh.
    â—¦ TÃ¡ch biÃªn (Boundary Extraction): Äá»ƒ láº¥y ra Ä‘Æ°á»ng biÃªn Ä‘á»‘i tÆ°á»£ng (náº¿u báº¡n dÃ¹ng áº£nh nhá»‹ phÃ¢n lÃ m input).

### Code tiá»n xá»­ lÃ­ áº£nh

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os # Import os module to remove file

def load_image(image_path):
    """
    Äá»c áº£nh tá»« Ä‘Æ°á»ng dáº«n Ä‘Ã£ cho.

    Parameters:
    - image_path (str): ÄÆ°á»ng dáº«n Ä‘áº¿n file áº£nh.

    Returns:
    - image (numpy.ndarray): áº¢nh Ä‘Ã£ Ä‘Æ°á»£c Ä‘á»c dÆ°á»›i dáº¡ng máº£ng NumPy (BGR).
    - None: Náº¿u khÃ´ng thá»ƒ Ä‘á»c áº£nh.
    """
    image = cv2.imread(image_path)
    if image is None:
        print(f"Lá»—i: KhÃ´ng thá»ƒ Ä‘á»c áº£nh tá»« Ä‘Æ°á»ng dáº«n: {image_path}")
    return image
def apply_histogram_equalization(image):
    """
    CÃ¢n báº±ng histogram Ä‘á»ƒ tÄƒng Ä‘á»™ tÆ°Æ¡ng pháº£n cá»§a áº£nh.
    Hiá»‡u quáº£ vá»›i áº£nh cÃ³ Ä‘á»™ tÆ°Æ¡ng pháº£n tháº¥p.
    """
    # Chuyá»ƒn Ä‘á»•i sang áº£nh xÃ¡m náº¿u lÃ  áº£nh mÃ u
    if len(image.shape) == 3:
        # Chuyá»ƒn Ä‘á»•i sang khÃ´ng gian mÃ u YCrCb
        ycrcb_img = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
        # CÃ¢n báº±ng histogram cho kÃªnh Y (Ä‘á»™ sÃ¡ng)
        ycrcb_img[:, :, 0] = cv2.equalizeHist(ycrcb_img[:, :, 0])
        # Chuyá»ƒn Ä‘á»•i láº¡i khÃ´ng gian mÃ u BGR
        equalized_img = cv2.cvtColor(ycrcb_img, cv2.COLOR_YCrCb2BGR)
    else:
        equalized_img = cv2.equalizeHist(image)

    return equalized_img
def restore_from_noise(image):
    """
    KhÃ´i phá»¥c áº£nh bá»‹ nhiá»…u báº±ng bá»™ lá»c Non-Local Means.
    ÄÃ¢y lÃ  má»™t ká»¹ thuáº­t khÃ´i phá»¥c áº£nh tiÃªn tiáº¿n, loáº¡i bá» nhiá»…u
    trong khi váº«n giá»¯ Ä‘Æ°á»£c cÃ¡c chi tiáº¿t vÃ  Ä‘Æ°á»ng nÃ©t.
    """
    if len(image.shape) == 3:
        # KhÃ´i phá»¥c áº£nh mÃ u
        restored_image = cv2.fastNlMeansDenoisingColored(image, None, 10, 10, 7, 21)
    else:
        # KhÃ´i phá»¥c áº£nh xÃ¡m
        restored_image = cv2.fastNlMeansDenoising(image, None, 10, 7, 21)
    return restored_image
def convert_to_grayscale(image):
    """
    Chuyá»ƒn Ä‘á»•i áº£nh tá»« BGR sang áº£nh xÃ¡m (Grayscale).

    Parameters:
    - image (numpy.ndarray): áº¢nh Ä‘áº§u vÃ o (BGR).

    Returns:
    - grayscale_image (numpy.ndarray): áº¢nh xÃ¡m.
    """
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

def apply_median_filter(image, kernel_size=5):
    """
    Ãp dá»¥ng Bá»™ lá»c trung vá»‹ (Median Filter) Ä‘á»ƒ loáº¡i bá» nhiá»…u.

    Parameters:
    - image (numpy.ndarray): áº¢nh Ä‘áº§u vÃ o (nÃªn lÃ  áº£nh xÃ¡m).
    - kernel_size (int): KÃ­ch thÆ°á»›c cá»§a kernel (pháº£i lÃ  sá»‘ láº», vÃ­ dá»¥: 3, 5, 7).

    Returns:
    - denoised_image (numpy.ndarray):  áº£nh Ä‘Ã£ Ä‘Æ°á»£c lÃ m giáº£m nhiá»…u.
    """
    return cv2.medianBlur(image, kernel_size)

def apply_canny_edge_detection(image, lower_threshold=None, upper_threshold=None):
    """
    PhÃ¡t hiá»‡n biÃªn trong áº£nh sá»­ dá»¥ng thuáº­t toÃ¡n Canny.

    Parameters:
    - image (numpy.ndarray): áº¢nh Ä‘áº§u vÃ o (nÃªn lÃ  áº£nh xÃ¡m, Ä‘Ã£ giáº£m nhiá»…u).
    - lower_threshold (int, optional): NgÆ°á»¡ng dÆ°á»›i cho Canny.
    - upper_threshold (int, optional): NgÆ°á»¡ng trÃªn cho Canny.

    Returns:
    - edge_image (numpy.ndarray): áº¢nh biÃªn (áº£nh nhá»‹ phÃ¢n).
    """
    # Tá»± Ä‘á»™ng xÃ¡c Ä‘á»‹nh ngÆ°á»¡ng náº¿u khÃ´ng Ä‘Æ°á»£c cung cáº¥p
    if lower_threshold is None or upper_threshold is None:
        median = np.median(image)
        lower = int(max(0, 0.7 * median))
        upper = int(min(255, 1.3 * median))
    else:
        lower = lower_threshold
        upper = upper_threshold

    return cv2.Canny(image, lower, upper)
def segment_by_thresholding(image):
    """
    PhÃ¢n vÃ¹ng áº£nh báº±ng phÆ°Æ¡ng phÃ¡p ngÆ°á»¡ng Otsu.
    Tá»± Ä‘á»™ng tÃ¬m ngÆ°á»¡ng tá»‘i Æ°u Ä‘á»ƒ tÃ¡ch Ä‘á»‘i tÆ°á»£ng vÃ  ná»n.
    """
    if len(image.shape) == 3:
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray_image = image

    # Ãp dá»¥ng ngÆ°á»¡ng Otsu
    _, segmented_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return segmented_image

def apply_morphological_opening(image, kernel_size=(3, 3), iterations=1):
    """
    Ãp dá»¥ng phÃ©p Má»Ÿ (Morphological Opening) Ä‘á»ƒ loáº¡i bá» nhiá»…u nhá» vÃ  lÃ m má»‹n biÃªn.

    Parameters:
    - image (numpy.ndarray): áº¢nh Ä‘áº§u vÃ o (thÆ°á»ng lÃ  áº£nh biÃªn).
    - kernel_size (tuple): KÃ­ch thÆ°á»›c kernel.
    - iterations (int): Sá»‘ láº§n láº·p láº¡i phÃ©p toÃ¡n.

    Returns:
    - opened_image (numpy.ndarray): áº¢nh sau khi Ã¡p dá»¥ng phÃ©p má»Ÿ.
    """
    kernel = np.ones(kernel_size, np.uint8)
    return cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel, iterations=iterations)

def apply_morphological_closing(image, kernel_size=(3, 3), iterations=1):
    """
    Ãp dá»¥ng phÃ©p ÄÃ³ng (Morphological Closing) Ä‘á»ƒ káº¿t ná»‘i cÃ¡c Ä‘á»‘i tÆ°á»£ng vÃ  láº¥p Ä‘áº§y lá»— há»•ng.

    Parameters:
    - image (numpy.ndarray): áº¢nh Ä‘áº§u vÃ o.
    - kernel_size (tuple): KÃ­ch thÆ°á»›c kernel.
    - iterations (int): Sá»‘ láº§n láº·p láº¡i phÃ©p toÃ¡n.

    Returns:
    - closed_image (numpy.ndarray): áº¢nh sau khi Ã¡p dá»¥ng phÃ©p Ä‘Ã³ng.
    """
    kernel = np.ones(kernel_size, np.uint8)
    return cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel, iterations=iterations)

def extract_boundary(image, kernel_size=(3, 3)):
    """
    TÃ¡ch biÃªn cá»§a Ä‘á»‘i tÆ°á»£ng tá»« áº£nh nhá»‹ phÃ¢n.

    Parameters:
    - image (numpy.ndarray): áº¢nh nhá»‹ phÃ¢n Ä‘áº§u vÃ o.
    - kernel_size (tuple): KÃ­ch thÆ°á»›c kernel dÃ¹ng cho phÃ©p xÃ³i mÃ²n (erosion).

    Returns:
    - boundary_image (numpy.ndarray): áº¢nh chá»‰ chá»©a Ä‘Æ°á»ng biÃªn.
    """
    kernel = np.ones(kernel_size, np.uint8)
    eroded_image = cv2.erode(image, kernel)
    return cv2.subtract(image, eroded_image)

def display_processing_steps(original_image, steps):
    """
    Hiá»ƒn thá»‹ áº£nh gá»‘c vÃ  káº¿t quáº£ cá»§a tá»«ng bÆ°á»›c xá»­ lÃ½.

    Parameters:
    - original_image (numpy.ndarray): áº¢nh gá»‘c.
    - steps (dict): Má»™t dictionary chá»©a tÃªn vÃ  káº¿t quáº£ cá»§a cÃ¡c bÆ°á»›c.
    """
    num_steps = len(steps)
    plt.figure(figsize=(15, 10))

    # Hiá»ƒn thá»‹ áº£nh gá»‘c
    plt.subplot(2, 4, 1)
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    plt.title('1. áº¢nh gá»‘c')
    plt.axis('off')

    # Hiá»ƒn thá»‹ cÃ¡c bÆ°á»›c xá»­ lÃ½
    for i, (title, img) in enumerate(steps.items()):
        plt.subplot(2, 4, i + 2)
        cmap = 'gray' if len(img.shape) == 2 else None
        plt.imshow(img, cmap=cmap)
        plt.title(f'{i + 2}. {title}')
        plt.axis('off')

    plt.tight_layout()
    plt.show()

def full_preprocessing_pipeline(image_path, display_steps=True):
    """
    Thá»±c hiá»‡n toÃ n bá»™ quy trÃ¬nh tiá»n xá»­ lÃ½ áº£nh báº±ng cÃ¡ch gá»i cÃ¡c hÃ m riÃªng láº».

    Parameters:
    - image_path (str): ÄÆ°á»ng dáº«n Ä‘áº¿n file áº£nh.
    - display_steps (bool): CÃ³ hiá»ƒn thá»‹ cÃ¡c bÆ°á»›c xá»­ lÃ½ hay khÃ´ng.

    Returns:
    - grayscale_image (numpy.ndarray): áº¢nh xÃ¡m Ä‘Ã£ Ä‘Æ°á»£c xá»­ lÃ½.
    - final_image (numpy.ndarray): áº¢nh Ä‘Ã£ Ä‘Æ°á»£c xá»­ lÃ½ cuá»‘i cÃ¹ng, sáºµn sÃ ng cho CNN.
    - None: Náº¿u cÃ³ lá»—i xáº£y ra.
    """
    # BÆ°á»›c 0: Äá»c áº£nh
    original_image = load_image(image_path)
    if original_image is None:
        return None, None # Return two Nones if image loading fails
    original_image = cv2.resize(original_image, (100, 100))
    processed_results = {}

    # I. LÃ m sáº¡ch áº£nh vÃ  TÄƒng cÆ°á»ng cháº¥t lÆ°á»£ng

    # 1. Chuyá»ƒn áº£nh xÃ¡m
    grayscale_image = convert_to_grayscale(original_image)
    processed_results['áº¢nh xÃ¡m'] = grayscale_image
    #  CÃ¢n báº±ng histogram
    histogram_img = apply_histogram_equalization(grayscale_image)
    processed_results['CÃ¢n báº±ng histogram'] = histogram_img

    # 2. Loáº¡i bá» nhiá»…u báº±ng Median Filter
    denoised_image = apply_median_filter(grayscale_image)
    processed_results['Loáº¡i bá» nhiá»…u (Median)'] = denoised_image

    # II. TÃ¡ch Ä‘á»‘i tÆ°á»£ng (Segmentation)

    # 3. PhÃ¡t hiá»‡n biÃªn báº±ng Canny
    edge_image = apply_canny_edge_detection(denoised_image)
    processed_results['PhÃ¡t hiá»‡n biÃªn (Canny)'] = edge_image

    otsu_img = segment_by_thresholding(grayscale_image)
    processed_results['PhÆ°Æ¡ng phÃ¡p otsu'] = otsu_img

    # 5c. TÃ¡ch biÃªn (Boundary Extraction)
    # final_image = extract_boundary(edge_image)
    processed_results['áº¢nh Ä‘áº§u vÃ o cho CNN'] = edge_image

    if display_steps:
        display_processing_steps(original_image, processed_results)

    return grayscale_image, edge_image

    # Äáº§u ra lÃ  áº£nh dáº¡ng (height , with , 2 )
```

# TrÃ­ch xuáº¥t Ä‘áº·c trÆ°ng thá»§ cÃ´ng

```python
import cv2
import numpy as np
import pandas as pd
from skimage.feature import graycomatrix, graycoprops
import os
from tqdm import tqdm # ThÆ° viá»‡n Ä‘á»ƒ hiá»ƒn thá»‹ thanh tiáº¿n trÃ¬nh
import warnings

# --- Äá»‹nh nghÄ©a hÃ m trÃ­ch xuáº¥t Ä‘áº·c trÆ°ng ---

def extract_manual_features(image_gray):
    """
    TrÃ­ch xuáº¥t 10 Ä‘áº·c trÆ°ng thá»§ cÃ´ng tá»« má»™t áº£nh xÃ¡m (grayscale_image).
    Bao gá»“m cÃ¡c Ä‘áº·c trÆ°ng vá» HÃ¬nh dáº¡ng (Shape) vÃ  Káº¿t cáº¥u (Texture).
    """
    epsilon = 1e-6
    # Vector Ä‘áº·c trÆ°ng sáº½ cÃ³ 10 chiá»u:
    # (1 AspectRatio + 1 Circularity + 5 HuMoments + 3 GLCM)
    feature_vector = np.zeros(10)

    # --- 1. Tiá»n xá»­ lÃ½ áº£nh nhá»‹ phÃ¢n Ä‘á»ƒ tÃ¬m Ä‘Æ°á»ng bao ---
    _, image_thresh = cv2.threshold(image_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Äáº£m báº£o váº­t thá»ƒ luÃ´n lÃ  MÃ€U TRáº®NG (ná»n Ä‘en).
    # Náº¿u Otsu ra váº­t thá»ƒ Ä‘en (Ã­t pixel tráº¯ng hÆ¡n), ta Ä‘áº£o áº£nh.
    if np.sum(image_thresh == 255) < np.sum(image_thresh == 0):
        image_thresh = cv2.bitwise_not(image_thresh)

    # --- 2. LÃ m sáº¡ch Ä‘Æ°á»ng bao báº±ng Morphology ---
    # DÃ¹ng kernel hÃ¬nh elip/trÃ²n, phÃ¹ há»£p vá»›i váº­t thá»ƒ tá»± nhiÃªn.
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3,3))

    # MORPH_OPEN: Loáº¡i bá» nhiá»…u tráº¯ng (cÃ¡c cháº¥m tráº¯ng nhá») bÃªn ngoÃ i.
    image_thresh = cv2.morphologyEx(image_thresh, cv2.MORPH_OPEN, kernel, iterations=1)
    # MORPH_CLOSE: Láº¥p Ä‘áº§y cÃ¡c lá»— Ä‘en (cháº¥m Ä‘en nhá») bÃªn trong váº­t thá»ƒ.
    image_thresh = cv2.morphologyEx(image_thresh, cv2.MORPH_CLOSE, kernel, iterations=1)

    # --- 3. TÃ¬m Ä‘Æ°á»ng bao (Contour) ---
    contours, _ = cv2.findContours(image_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Náº¿u khÃ´ng tÃ¬m tháº¥y contour (áº£nh trá»‘ng), tráº£ vá» vector 0
    if not contours:
        return feature_vector

    # Chá»‰ láº¥y Ä‘Æ°á»ng bao lá»›n nháº¥t (giáº£ Ä‘á»‹nh lÃ  váº­t thá»ƒ chÃ­nh)
    main_contour = max(contours, key=cv2.contourArea)

    # --- 4. TÃ­nh Ä‘áº·c trÆ°ng HÃŒNH Dáº NG (Shape) ---

    # a. Aspect Ratio (Tá»‰ lá»‡ khung bao)
    x, y, w, h = cv2.boundingRect(main_contour)
    aspect_ratio = float(w) / (h + epsilon)

    # b. Circularity (Äá»™ trÃ²n)
    area = cv2.contourArea(main_contour)
    perimeter = cv2.arcLength(main_contour, True)
    circularity = (4 * np.pi * area) / ((perimeter * perimeter) + epsilon)

    # c. Hu Moments (Láº¥y 5 giÃ¡ trá»‹ Ä‘áº§u)
    moments = cv2.moments(main_contour)
    hu_moments_all = cv2.HuMoments(moments).flatten()
    hu_moments = hu_moments_all[:5] # Chá»‰ láº¥y 5 giÃ¡ trá»‹ Ä‘áº§u, bá» 2 giÃ¡ trá»‹ cuá»‘i dá»… nhiá»…u

    # Log-transform Ä‘á»ƒ á»•n Ä‘á»‹nh thang Ä‘o cá»§a Hu moments
    for i in range(5):
        hu_moments[i] = -1 * np.sign(hu_moments[i]) * np.log10(np.abs(hu_moments[i]) + epsilon)

    # --- 5. TÃ­nh Ä‘áº·c trÆ°ng Káº¾T Cáº¤U (Texture) ---

    # LÆ°á»£ng tá»­ hÃ³a áº£nh xuá»‘ng 8 má»©c xÃ¡m (0-7) Ä‘á»ƒ tÃ­nh GLCM hiá»‡u quáº£
    image_quantized = (image_gray // 32).astype(np.uint8)

    # TÃ­nh GLCM (Tá»‘i Æ°u: 1 hÆ°á»›ng, 1 khoáº£ng cÃ¡ch)
    glcm = graycomatrix(image_quantized,
                        distances=[1], # Khoáº£ng cÃ¡ch 1 pixel
                        angles=[0],    # GÃ³c 0 Ä‘á»™
                        levels=8,
                        symmetric=True,
                        normed=True)

    # TrÃ­ch xuáº¥t 3 Ä‘áº·c trÆ°ng tá»« GLCM
    contrast = graycoprops(glcm, 'contrast')[0, 0]
    homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]
    energy = graycoprops(glcm, 'energy')[0, 0]

    # --- 6. Tá»•ng há»£p thÃ nh vector cuá»‘i cÃ¹ng (10 chiá»u) ---
    feature_vector = np.array([
        aspect_ratio,     # 0
        circularity,      # 1
        hu_moments[0],    # 2
        hu_moments[1],    # 3
        hu_moments[2],    # 4
        hu_moments[3],    # 5
        hu_moments[4],    # 6
        contrast,         # 7 (Káº¿t cáº¥u)
        homogeneity,      # 8 (Káº¿t cáº¥u)
        energy            # 9 (Káº¿t cáº¥u)
    ])

    # Xá»­ lÃ½ cÃ¡c giÃ¡ trá»‹ NaN/Inf (náº¿u cÃ³)
    feature_vector = np.nan_to_num(feature_vector, nan=0.0, posinf=0.0, neginf=0.0)

    # Giá»›i háº¡n giÃ¡ trá»‹ trong khoáº£ng an toÃ n [-50, 50] Ä‘á»ƒ á»•n Ä‘á»‹nh khi huáº¥n luyá»‡n
    feature_vector = np.clip(feature_vector, -50.0, 50.0)

    return feature_vector

# --- Báº¯t Ä‘áº§u cháº¡y trÃ­ch xuáº¥t trÃªn toÃ n bá»™ dá»¯ liá»‡u ---

print("Báº¯t Ä‘áº§u trÃ­ch xuáº¥t Ä‘áº·c trÆ°ng thá»§ cÃ´ng (manual features)...")

# Giáº£ Ä‘á»‹nh 'image_files' vÃ  'full_preprocessing_pipeline' Ä‘Ã£ tá»“n táº¡i tá»« cell trÆ°á»›c

all_features = []
all_image_paths = []
output_feature_file = 'manual_features.csv'

warnings.filterwarnings('ignore') # Táº¡m táº¯t cáº£nh bÃ¡o cá»§a skimage/cv2

# Duyá»‡t qua táº¥t cáº£ cÃ¡c file áº£nh vÃ  trÃ­ch xuáº¥t Ä‘áº·c trÆ°ng
for img_path in tqdm(image_files, desc="Äang trÃ­ch xuáº¥t Ä‘áº·c trÆ°ng"):

    # Chá»‰ cáº§n áº£nh xÃ¡m (grayscale_img) Ä‘á»ƒ tÃ­nh Ä‘áº·c trÆ°ng
    grayscale_img, _ = full_preprocessing_pipeline(img_path, display_steps=False)

    if grayscale_img is not None:
        # Gá»i hÃ m trÃ­ch xuáº¥t
        feature_vector = extract_manual_features(grayscale_img)
        all_features.append(feature_vector)
        all_image_paths.append(img_path) # LÆ°u Ä‘Æ°á»ng dáº«n Ä‘á»ƒ khá»›p dá»¯ liá»‡u
    else:
        print(f"Bá» qua (lá»—i pipeline): {img_path}")

warnings.filterwarnings('default') # Báº­t láº¡i cáº£nh bÃ¡o

# --- LÆ°u táº¥t cáº£ Ä‘áº·c trÆ°ng vÃ o 1 file CSV duy nháº¥t ---
print(f"\nÄÃ£ trÃ­ch xuáº¥t hoÃ n táº¥t. Tá»•ng cá»™ng {len(all_features)} vector.")
print("Äang lÆ°u vÃ o file CSV...")

# Äáº·t tÃªn 10 cá»™t Ä‘áº·c trÆ°ng
column_names = [
    'aspect_ratio', 'circularity',
    'hu1', 'hu2', 'hu3', 'hu4', 'hu5',
    'contrast', 'homogeneity', 'energy'
]

# Táº¡o DataFrame
df_features = pd.DataFrame(all_features, columns=column_names)

# ChÃ¨n cá»™t 'image_path' vÃ o Ä‘áº§u tiÃªn Ä‘á»ƒ dá»… dÃ ng map vá»›i file .npy
df_features.insert(0, 'image_path', all_image_paths)

# LÆ°u ra file CSV
df_features.to_csv(output_feature_file, index=False)

print(f"ÄÃ£ lÆ°u thÃ nh cÃ´ng vÃ o: {output_feature_file}")

# Hiá»ƒn thá»‹ 5 dÃ²ng Ä‘áº§u tiÃªn cá»§a file CSV Ä‘á»ƒ kiá»ƒm tra
print(f"\n5 dÃ²ng Ä‘áº§u tiÃªn cá»§a file Ä‘áº·c trÆ°ng ({output_feature_file}):")
print(df_features.head())
```

```python
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# 1. Load dá»¯ liá»‡u vÃ  nhÃ£n
X = []
y = []

print("Äang táº£i dá»¯ liá»‡u tá»« cÃ¡c file .npy...")

# processed_data_for_cnn lÃ  list chá»©a [(original_path, npy_path), ...] tá»« code trÆ°á»›c cá»§a báº¡n
# Náº¿u báº¡n Ä‘Ã£ cháº¡y xong code trÆ°á»›c, biáº¿n nÃ y váº«n cÃ²n trong bá»™ nhá»›.
for original_path, npy_path in processed_data_for_cnn:
    try:
        # Load áº£nh 2 kÃªnh
        data = np.load(npy_path)
        X.append(data)

        # Láº¥y nhÃ£n tá»« tÃªn thÆ° má»¥c cha (vÃ­ dá»¥: .../Apple/apple01.jpg -> label: Apple)
        label = os.path.basename(os.path.dirname(original_path))
        y.append(label)
    except Exception as e:
        print(f"Lá»—i khi Ä‘á»c file {npy_path}: {e}")

X = np.array(X)
y = np.array(y)

# Chuáº©n hÃ³a dá»¯ liá»‡u vá» khoáº£ng [0, 1] Ä‘á»ƒ mÃ´ hÃ¬nh há»c tá»‘t hÆ¡n
X = X.astype('float32') / 255.0

print(f"KÃ­ch thÆ°á»›c dá»¯ liá»‡u Ä‘áº§u vÃ o (X): {X.shape}") # (Sá»‘ lÆ°á»£ng áº£nh, 100, 100, 2)
print(f"Sá»‘ lÆ°á»£ng nhÃ£n (y): {len(y)}")

# 2. MÃ£ hÃ³a nhÃ£n (Label Encoding)
le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_categorical = to_categorical(y_encoded) # One-hot encoding
num_classes = len(le.classes_)

print("CÃ¡c loáº¡i quáº£ Ä‘Ã£ tÃ¬m tháº¥y:", le.classes_)

# 3. Chia táº­p train vÃ  test (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.2, random_state=42)

print(f"Training set: {X_train.shape}")
print(f"Test set: {X_test.shape}")



import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Input

def build_cnn_model(input_shape, num_classes):
    model = Sequential()

    # Input layer: ChÃº Ã½ input_shape lÃ  (100, 100, 2)
    model.add(Input(shape=input_shape))

    # Block 1: TrÃ­ch xuáº¥t Ä‘áº·c trÆ°ng cáº¥p tháº¥p
    model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D((2, 2)))

    # Block 2: TrÃ­ch xuáº¥t Ä‘áº·c trÆ°ng cáº¥p trung
    model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D((2, 2)))
    model.add(Dropout(0.25)) # TrÃ¡nh overfitting

    # # Block 3: TrÃ­ch xuáº¥t Ä‘áº·c trÆ°ng cáº¥p cao
    # model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
    # model.add(BatchNormalization())
    # model.add(MaxPooling2D((2, 2)))
    # model.add(Dropout(0.25))

    # Flatten: Duá»—i tháº³ng Ä‘á»ƒ Ä‘Æ°a vÃ o máº¡ng nÆ¡-ron Ä‘áº§y Ä‘á»§ (Dense)
    model.add(Flatten())

    # Fully Connected Layers (PhÃ¢n lá»›p)
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax')) # Output layer

    return model

# Khá»Ÿi táº¡o model
model = build_cnn_model((100, 100, 2), num_classes)
model.summary()

# Compile model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])



import matplotlib.pyplot as plt

# Huáº¥n luyá»‡n
EPOCHS = 20 # Báº¡n cÃ³ thá»ƒ tÄƒng sá»‘ epoch náº¿u cáº§n
BATCH_SIZE = 32

history = model.fit(
    X_train, y_train,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_data=(X_test, y_test),
    verbose=1
)

# --- Váº½ biá»ƒu Ä‘á»“ Accuracy vÃ  Loss ---
def plot_training_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(len(acc))

    plt.figure(figsize=(12, 5))

    # Biá»ƒu Ä‘á»“ Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')
    plt.grid(True)

    # Biá»ƒu Ä‘á»“ Loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')
    plt.grid(True)

    plt.show()

plot_training_history(history)

# LÆ°u model
model.save('fruit_recognition_cnn_2channel.h5')
print("ÄÃ£ lÆ°u mÃ´ hÃ¬nh thÃ nh cÃ´ng!")
```

```python
  def predict_fruit(image_path, model, label_encoder):
    """
    HÃ m dá»± Ä‘oÃ¡n loáº¡i quáº£ tá»« Ä‘Æ°á»ng dáº«n áº£nh, hiá»ƒn thá»‹ 5 dá»± Ä‘oÃ¡n hÃ ng Ä‘áº§u
    vÃ  dá»± Ä‘oÃ¡n cuá»‘i cÃ¹ng (cÃ³ Ä‘á»™ tin cáº­y cao nháº¥t).
    """
    # 1. Tiá»n xá»­ lÃ½ áº£nh (Sá»­ dá»¥ng láº¡i pipeline cá»§a báº¡n)
    grayscale_img, edge_img = full_preprocessing_pipeline(image_path, display_steps=False) # display_steps=False Ä‘á»ƒ trÃ¡nh hiá»ƒn thá»‹ quÃ¡ nhiá»u biá»ƒu Ä‘á»“

    if grayscale_img is None:
        print("KhÃ´ng thá»ƒ Ä‘á»c áº£nh.")
        return

    # 2. Resize (Ä‘áº£m báº£o Ä‘Ãºng kÃ­ch thÆ°á»›c 100x100)
    if grayscale_img.shape != (100, 100):
        grayscale_img = cv2.resize(grayscale_img, (100, 100))
        edge_img = cv2.resize(edge_img, (100, 100))

    # 3. Táº¡o Ä‘á»‹nh dáº¡ng 2 kÃªnh (Height, Width, 2)
    grayscale_chan = np.expand_dims(grayscale_img, axis=-1)
    edge_chan = np.expand_dims(edge_img, axis=-1)
    input_img = np.concatenate([grayscale_chan, edge_chan], axis=-1)

    # 4. Chuáº©n hÃ³a (chia 255.0 giá»‘ng lÃºc train)
    input_img = input_img.astype('float32') / 255.0

    # 5. Má»Ÿ rá»™ng chiá»u batch (1, 100, 100, 2)
    input_batch = np.expand_dims(input_img, axis=0)

    # 6. Dá»± Ä‘oÃ¡n
    predictions = model.predict(input_batch)
    probabilities = predictions[0] # Láº¥y máº£ng xÃ¡c suáº¥t cho má»™t áº£nh Ä‘áº§u vÃ o

    # Sáº¯p xáº¿p cÃ¡c xÃ¡c suáº¥t vÃ  láº¥y chá»‰ sá»‘ cá»§a 5 dá»± Ä‘oÃ¡n hÃ ng Ä‘áº§u
    top_5_indices = np.argsort(probabilities)[::-1][:5]
    top_5_probabilities = probabilities[top_5_indices]

    # Láº¥y tÃªn lá»›p tÆ°Æ¡ng á»©ng cho 5 dá»± Ä‘oÃ¡n hÃ ng Ä‘áº§u
    top_5_labels = label_encoder.inverse_transform(top_5_indices)

    # Dá»± Ä‘oÃ¡n cuá»‘i cÃ¹ng (cÃ³ Ä‘á»™ tin cáº­y cao nháº¥t)
    predicted_class_idx = np.argmax(probabilities)
    predicted_label = label_encoder.inverse_transform([predicted_class_idx])[0]
    confidence = np.max(probabilities) * 100

    # 7. Hiá»ƒn thá»‹ káº¿t quáº£
    plt.figure(figsize=(15, 6))

    # Hiá»ƒn thá»‹ áº£nh gá»‘c (Ä‘á»c láº¡i Ä‘á»ƒ hiá»ƒn thá»‹ mÃ u)
    original_bgr = cv2.imread(image_path)
    original_rgb = cv2.cvtColor(original_bgr, cv2.COLOR_BGR2RGB)

    plt.subplot(1, 2, 1)
    plt.imshow(original_rgb)
    plt.title(f"Input: {os.path.basename(image_path)}")
    plt.axis('off')

    # Hiá»ƒn thá»‹ káº¿t quáº£ dá»± Ä‘oÃ¡n
    plt.subplot(1, 2, 2)
    display_text = f"Dá»± Ä‘oÃ¡n cuá»‘i cÃ¹ng: {predicted_label} ({confidence:.2f}%)\n\nTop 5 Dá»± Ä‘oÃ¡n:\n"
    for i in range(len(top_5_labels)):
        display_text += f"- {top_5_labels[i]}: {top_5_probabilities[i]*100:.2f}%\n"

    plt.text(0.1, 0.5, display_text, fontsize=12, verticalalignment='center', horizontalalignment='left')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

    return predicted_label, confidence

# --- CHáº Y THá»¬ NGHIá»†M ---
# Láº¥y má»™t áº£nh báº¥t ká»³ tá»« táº­p test hoáº·c áº£nh ngoÃ i Ä‘á»ƒ thá»­
# LÆ°u Ã½: Thay Ä‘á»•i Ä‘Æ°á»ng dáº«n nÃ y Ä‘áº¿n áº£nh báº¡n muá»‘n dá»± Ä‘oÃ¡n
# duong_dan = "/root/.cache/kagglehub/datasets/sshikamaru/fruit-recognition/versions/2/test/test/Apple Braeburn_0.jpg" # VÃ­ dá»¥ áº£nh Apple Braeburn
duong_dan = "/kaggle/input/fruit-recognition/train/train/Orange/Orange_4.jpg" # VÃ­ dá»¥ áº£nh khÃ´ng pháº£i quáº£
# duong_dan = "/kaggle/input/fruit-recognition/test/test/0001.jpg"

print(f"Äang dá»± Ä‘oÃ¡n áº£nh: {duong_dan}")

pred_label, conf = predict_fruit(duong_dan, model, le)
```
